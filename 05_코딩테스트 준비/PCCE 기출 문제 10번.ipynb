 - .sort() — 리스트를 정렬하는 메서드
- key= — 정렬 기준을 지정하는 파라미터
- lambda x: x[d_sort_by] — 익명 함수 (이름 없는 함수)


```python
# 일반 함수
def double(x):
    return x * 2

# lambda로 같은 것
double = lambda x: x * 2

print(double(5))  # → 10
```

```python

nums = [5, -3, 1, -8, 2]

# 절대값 기준으로 정렬
nums.sort(key=lambda x: abs(x))
print(nums)  # → [1, 2, -3, 5, -8]
```

```python
pythonwords = ["banana", "pie", "apple", "go"]

words.sort(key=lambda x: len(x))
print(words)  # → ['go', 'pie', 'apple', 'banana']
```

```python
pythonstudents = [("철수", 85), ("영희", 92), ("민수", 78)]

students.sort(key=lambda x: x[1])
print(students)
# → [('민수', 78), ('철수', 85), ('영희', 92)]
```

```python
items = [
    {"name": "사과", "price": 3000},
    {"name": "바나나", "price": 1500},
    {"name": "포도", "price": 5000}
]

items.sort(key=lambda x: x["price"])
print(items)
# → [바나나(1500), 사과(3000), 포도(5000)]
```

```python
def solution(data, ext, val_ext, sort_by):
    order_data = {
        'code': 0,
        'date': 1,
        'maximum': 2,
        'remain': 3
    }

    d_ext = order_data[ext]
    d_sort_by = order_data[sort_by]

    result = []

    for i in range(len(data)):
        if data[i][d_ext] < val_ext:
            result.append(data[i])

    result.sort(key=lambda x: x[d_sort_by])

    return result
```
