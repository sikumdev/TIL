```python
def solution(new_id):
    # step 1
    new_id = new_id.lower()

    # step 2
    for c in new_id:
        if c.isalnum() or c in '._-':
            pass
        else:
            new_id = new_id.replace(c,'')
    
    # step 3
    while '..' in new_id:
        new_id = new_id.replace('..','.')

    # step 4

    new_id = new_id.strip('.')

    # step 5

    if len(new_id) == 0:
        new_id += 'a'

    # step 6
    if len(new_id) >= 16:
        new_id = new_id[:15]
        new_id = new_id.strip('.')

    # step 7
    if len(new_id) <= 2:
        while len(new_id) < 3:
            new_id += new_id[-1]

    result = new_id
    return result
```
