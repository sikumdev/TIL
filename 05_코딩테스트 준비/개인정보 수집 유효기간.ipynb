- map(자료형,값들)
- .split(기준)

```python
def solution(today, terms, privacies):
    y,m,d = map(int,today.split('.'))
    total_days = (y * 28 * 12) + (m * 28) + d
    
    terms_table = {}
    for t in terms:
        kind, duration = t.split()
        terms_table[kind] = int(duration)
    
    result = []
    i = 0
    for p in privacies:
        start_date,kind  = p.split()
        year,month,date = map(int,start_date.split('.'))
        start_days = year * 12 * 28 + month * 28 + date
        if total_days >= start_days + terms_table[kind] * 28:
            result.append(i+1)
        i += 1
        
    return result
        
```
