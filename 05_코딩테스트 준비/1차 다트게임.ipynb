- .replace('10','N') -> 10을 바로 처리해야함
- .isalpha() -> 알파벳만 가능 (특수문자 X)

```python
def solution(dartResult):

    score = []
    idx = 0
    dartResult.replace('10','N')
    while idx < len(dartResult):
        if dartResult[idx].isdigit():
            if dartResult[idx] == '1' and dartResult[idx+1] =='0':
                score.append(10)
                idx+=1
            else:
                score.append(int(dartResult[idx]))
             
        
        elif dartResult[idx].isalpha():
            if dartResult[idx] == 'D':
                score[-1] = score[-1] ** 2
             
            elif dartResult[idx] == 'T':
                score[-1] = score[-1] ** 3
                
        
        else:
            if dartResult[idx] == '*' :
                if len(score)-2 >= 0:
                    score[len(score)-1] *=2 
                    score[len(score)-2] *=2 
                else:
                    score[len(score)-1] *=2 

            elif dartResult[idx] == '#':
                score[len(score)-1] *= -1
        
        idx+=1
        
    return sum(score)
    
```
