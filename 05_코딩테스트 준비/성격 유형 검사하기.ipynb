```python
def solution(survey, choices):
    
    surv= {'R':0, 'T':0,'C':0, 'F':0,'J':0, 'M':0,'A':0, 'N':0}
    type_s = [['R','T'],['C','F'],['J','M'],['A','N']]
    
    for i,v in enumerate(survey):
        if choices[i] < 4:
            surv[v[0]] += 4 - choices[i]
        elif choices[i] >4:
            surv[v[1]] += choices[i] - 4
        
    answer = ''

    for t in type_s:
        if surv[t[0]] > surv[t[1]]:
            answer += t[0]
        elif surv[t[0]] < surv[t[1]]:
            answer += t[1]
        else:
            if t[0] < t[1]:
                answer += t[0]
            else:
                answer += t[1]
    return answer
```
