```python
def solution(s):
    
    
    words = ["zero","one","two","three","four","five","six","seven","eight","nine"]
    
    buffer = ''
    answer = ''
    for i,c in enumerate(s):
        if c.isalpha():
            buffer += c
        else:
            answer += c
            
        if buffer in words:
            answer = s.replace(buffer,str(words.index(buffer))) # <- 인덱스값을 어떻게 찾지?
            buffer = ''
    
    return int(s)
    
```

```python
def solution(s):
    words = {"zero":"0","one":"1","two":"2","three":"3","four":"4",
             "five":"5","six":"6","seven":"7","eight":"8","nine":"9"}
    
    result = ''
    buffer = ''

    for c in s:
        if c.isdigit():
            result += c
        else:
            buffer += c
            if buffer in words:
                result += words[buffer]
                buffer = ''
    
    return int(result)
```
