### Short-Circuit Evaluation (단축 평가)
- 앞에서 값이 확실 시 되면, 그 뒤는 확인하지 않는다
- 조건 판별 시 뒷부분을 확인하지 않아도 됨.

```python
# and 연산은 하나라도 False이면 결과값은 False 임 
a = True and True and True and False and True and False
# ? -> ? -> ? ......  !     [종료, 뒤에 볼 필요 없음]

# or 연산은 하나라도 True이면 결과값은 True임
b = False or False or True or False
# ? -> ? - ......  !     [종료, 뒤에 볼 필요 없음]

print(a,b)
```

```python
x = 1 and 2 and 3 and 0 and 4 # 결과값은 0이 나옴 (마지막으로 본 결과값을 출력함)
xx = True and 1
y = 0 or None or '' or [] or [None] or 10 or 'asdf'
print(x,xx, y)
```

```python
# bmi 계산기
# bmi = 몸무게(kg) / 키(m) 제곱 2
# weight 에 몸무게, height 에 키를 저장하고 bmi 를 계산하기

weight = 44
height = 1.55
bmi = weight / (height ** 2)
print(f'몸무게 : {weight}kg, 키 :{height*100}cm, bmi : {bmi} ')
```

```python
# 원 넓이와 둘레 계산 (pi 는 3.14 로 계산)
# 원 넓이 = pi x r^2
# 원 둘레 = 2 x pi x r

import math
r = 3
area = math.pi * (r ** 2)
perimeter= 2 * math.pi * r
print(f'원 넓이: {area}, 원 둘레: {perimeter}')
```

- 계산을 통해 h, m, s 에 시간, 분, 초로 전환

```python
seconds = 3675  # 1시간 1분 15초
s = seconds % 60
m = (seconds // 60) % 60
h = (seconds//60) // 60

print(f'{h}시간 {m}분 {s}초')

# s,m,h

```

```python
seconds = 3675  # 1시간 1분 15초

# 시간을 구하고 
h = seconds // (60 * 60)
# 그만큼 뺀다
seconds = seconds - (h * 60 * 60)

# 분을 구하고
m = seconds // 60 
# 그만큼 뺀다
seconds = seconds - (m * 60)

# 초를 구한다
s = seconds

h,m,s
```

```python
seconds = 3675  # 1시간 1분 15초

h, seconds = divmod(seconds, 3600) # -> 몫과 나머지가 나옴
m, seconds = divmod(seconds,60) 
s = seconds

print(f'{h}시간 {m}분 {s}초')
```

- 숫자를 백, 십, 일의 자리로 분리하여 변수에 저장하기

```python
n = 472

first = n // 100
second = n % 100 // 10
last = n % 100 % 10
first, second , last

print(f'백의 자리:{first}, 십의 자리:{second}, 일의 자리:{last}')
```

```python
n = 472
num_str = str(n)
num_digit_list = []

for n in num_str:
    num_digit_list.append(int(n))


print(num_digit_list)
```

```python
n = 472
ns = str(n)
num_digit_list =[]

for num in range(1, len(ns)+1):
    num_digit_list.append(n // 10**(len(ns)-num))
    n %= 10**(len(ns)-num)
    
print(num_digit_list)
```
