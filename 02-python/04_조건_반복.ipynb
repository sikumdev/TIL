# 04_조건_반복

```python
# 조건 - if

n = -10

if n > 0 :
    # 블럭 안은 조건이 True 일때만 실행
    print('양수')
    print('111111')

# 블럭 밖은 조건문과 상관 없음
print('22222')
```

```python
# 조건 -if-else
n = 0

if n > 0 :
    # 블럭 안은 조건이 True 일때만 실행
    print('양수')
    print('111111')

else : 
    print('0보다 작거나 같음')
```

```python
# 조건 - if-elif-else
n = 10

if n > 0 :
    print('양수')

elif n == 0:
    print('0이다')

else : 
    print('음수')
```

```python
score = 599

# no level : 0-399
# 1 : 400-599
# 2 : 600-749
# 3 : 750-899
# 4 : 900-1000

if score > 899 :
    print('level4 입니다.')

elif score >749:
    print('level3 입니다.')

elif score >599:
    print('level2 입니다.')

elif score >399:
    print('level1 입니다.')

else:
    print('no level 입니다.')
```

```python
# 우수 학생 판별
status = {
    'score' : 80,
    'absent': 3
}

# 1. 점수가 70을 넘는 학생이면 우수 학생
# 2. 우수 학생 중에 결석일이 5일 미만이면 최 우수 학생
# 3. (우수 아닌) 결석일이 5일 미만이면 성실학생
# 4. 나머지는 일반 학생

if status['score'] > 70:
    if status['absent'] < 5:
        print('최우수')
    else : 
        print('우수')
else:
    if status['absent'] < 5:
        print('성실 학생')
    else : 
        print('일반 학생')
```

## `while` 구문

```python
n = 0
while n < 5:
    print('FIVE TIMES')
    n += 1

print('끝나고', n)
```

```python
n = 0
while n < 5:
    print('출력',n)
    n += 2

print('끝나고', n)
```

```python
# 
numbers = [1, 2, 3, 4, 5]

total = 0
idx = 0

while idx < len(numbers) :
    total += numbers[idx]
    print(f'인덱스 {idx}일때 -> total: {total}')
    idx += 1

print(f'total : {total}')
```

# `for .. in` 구문

```python
numbers = [1, 2, 3, 4, 5]
total = 0

for num in numbers:
    total += num

print(total)
```

```python
s = range(5)
print(s)
```

```python
# print 하면 시간이 엄청 오래걸려서 print없이 1000개 테스트 해봄
total = 0
for n in range(1000):
    total += n

print(total)
```
