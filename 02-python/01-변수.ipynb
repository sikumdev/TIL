## f-string(`f''`, `f''' '''`)
### String 안에 변수를 녹여서 써야할 때 사용

```python
name ='김시연'
age = 27
# 김시연님 환영합니다. 김시연님의 등급은 일반 입니다.
# greeting = name +'님 환영합니다 ' + name + '님의 등급은 ~~입니다.'

# f-string ( f'' )
greeting = f'{name}님 환영합니다.\n{name}님의 나이는 {age}살 입니다.'
print(greeting)
```

```python
# String 안에서 따옴표 사용하기
# siyeon's bag 

title = "siyeon's bag"
title = 'siyeon\'s bag'
story = '그는 말했다 "안녕?" 이라고.'
story = '그는 말했다. \'안녕?\'이라고'

print(title, story)
```
