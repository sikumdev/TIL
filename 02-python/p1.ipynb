```python
# 문제 1. 점수 평균과 합격 판별
scores = [80, 75, 90, 60, 100]
total_score = 0

for sc in scores:
    total_score+=sc

avg_scores = total_score / len(scores)


if avg_scores >= 80:
    print(f"평균 점수 : {avg_scores}, 합격")
else:
    print(f"평균 점수 : {avg_scores}, 불합격")
```

```python
# 문제 2. 최대값 찾기
numbers = [10, 7, 2, 9, 5]

# 우선 최대값 후보로 첫번째 숫자를 잡는다.
num = numbers[0]

for i in range(1,len(numbers)):
    if num < numbers[i]:
        num = numbers[i]

print(f'가장 큰 값: {num}')
        
```

```python
# 문제 7. 리스트 안 딕셔너리 평균 구하기
students = [
    {"name": "철수", "score": 80},
    {"name": "영희", "score": 90},
    {"name": "민수", "score": 70}
]

total_score = 0
avg_scores = 0

for student in students:
    total_score+=student["score"]

avg_scores = total_score / len(students)
print(f"전체 평균 점수는 {avg_scores}")

for student in students:
   if student["score"] >= avg_scores :
       print(f"{student["name"]}: {student["score"]}")
```

```python
# 문제 8. 등급 매기기

scores = [95, 82, 67, 74, 88]

grades = []

grade = ''

for sc in scores:
    if sc >= 90:
        grade ='A'
    elif sc >= 80:
        grade ='B'
    elif sc >= 70:
        grade ='C'
    else: 
        grade ='D'
    
    grades.append(grade)

print(grades)
```

```python
# 문제 9. 장바구니 총 금액 계산
cart = [
    {"item": "노트북", "price": 1000000, "qty": 1},
    {"item": "마우스", "price": 20000, "qty": 2}
]

total_amount = 0

for item in cart:
    total_amount += (item["price"] * item["qty"])

if total_amount >= 1000000:   # 1000000 대신 1e6 --> 1* 10^6
    print(f'총구매금액: {total_amount}, 고액 구매자')
else: 
    print(f'총구매금액: {total_amount}, 일반 구매자')
```

```python
from collections import deque

def solution(n,edges):
    graph = [[] for _ in range(n+1)]

    for a, b in edges:
        graph[a].append(b)
        graph[b].append(a)

    visited = [False] * (n + 1)
    queue = deque([1])
    visited[1] = True

    count = 0
    while queue :
        node = queue.popleft()
        count += 1
        for neighbor in graph[node]:
            if not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)
        
        return count
                
```

```python
from collections import deque

def solution(n, edges):
    # 1. 인접 리스트 만들기 (양방향)
    graph = [[] for _ in range(n + 1)]
    for a, b in edges:
        graph[a].append(b)
        graph[b].append(a)

    # 2. BFS
    visited = [False] * (n + 1)
    queue = deque([1])
    visited[1] = True

    count = 0
    while queue:
        node = queue.popleft()
        count += 1
        for neighbor in graph[node]:
            if not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)

    return count
```
