````python
# name : plus_one / in : 숫자 / out: 숫자 + 1

# 함수 정의 (define)
def plus_one(num):
    return (num+1)

# 함수 호출(call)
plus_one(9)
````

````python
# n번 인사하는 함수
# name : hello_ntimes

''' 
def hello_ntimes(n):
    return 'hello' * n
'''

def hello_ntimes(n):
    # for-매개변수에 들어간 값을 사용하지 않을 때 => _
    for _ in range(n):
        print('반가워')

hello_ntimes(5)
````

````python
# 3제곱 해주는 함수
# name : cube, in : 숫자 1개, out : 숫자의 세제곱

'''
def cube(num):
    answer = 1
    for _ in range(3):
        answer*=num
    return answer

cube(11)
'''

def cube(num):
    return num ** 3

cube(3)
````

````python
# 원기둥의 부피를 구하는 함수 -> 부피 공식 : pi * r^2 *h
# name: cylinder, in: 반지름/높이, out: 부피
import math

def cylinder(r,h):
    return (math.pi * r**2 * h)

cylinder(2,5)
````

````python
# 리스트 내부의 숫자들의 평균을 구해주는 함수
# name: get_avg, in 숫자들로 만들어진 리스트, out: 평균

def get_avg(li):
    total = 0
    for num in li:
        total+=num
    
    return total/len(li)


get_avg([1,2,3,4,5,6,7])
````

````python
# 비밀번호 체크해주는 함수 (숫자 포함, 8글자 이상)
# name : check_pw, in : 비밀번호 문자열, out: True/False

def check_pw(ch):
    flag = False

    if len(ch)>=8:
        for c in ch:
            if c.isdigit():
                flag = True
                break
    
    return flag

print(check_pw('abcdefg'))
print(check_pw('abcd1234'))
````
